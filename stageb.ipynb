```python
#data normilization
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
normalized_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
```

```python
features_df= normalized_df.drop(columns=['Temperature_in_kitchen','Appliances'])
```

```python
Appliances_target=normalized_df['Appliances']
```

```python
#tarining and testing dataset
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test= train_test_split(features_df, Appliances_target,train_size=0.7,test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae=mean_absolute_error(y_test,predicted_values)
round(mae,3)
```

```python
import numpy as np
rss=np.sum(np.square(y_test - predicted_values))
round(rss,3)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse=np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse,3)
```

```python
from sklearn.metrics import r2_score 
r2_score   =   r2_score(y_test,   predicted_values) 
round(r2_score,3 ) 
```

```python
from sklearn.linear_model import Ridge 
ridge_reg   =   Ridge(alpha= 0.5 ) 
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import  Lasso 
lasso_reg  =  Lasso(alpha= 0.001 ) 
lasso_reg.fit(x_train, y_train)
```

```python
def get_weights_df(model,   feat,   col_name) : 
    weights = pd.Series(model.coef_,feat.columns).sort_values() 
    weights_df = pd.DataFrame(weights).reset_index() 
    weights_df.columns = [ 'Features' ,  col_name] 
    weights_df[col_name].round( 3 ) 
    return weights_df 
 
 
```

```python
linear_model_weights = get_weights_df(linear_model,x_train,'Linear_Model_Weight' ) 
ridge_weights_df = get_weights_df(ridge_reg, features_df,'Ridge_Weight' ) 
lasso_weights_df =  get_weights_df(lasso_reg, features_df, 'Lasso_weight' ) 
final_weights   =   pd.merge(linear_model_weights,  ridge_weights_df,   on= 'Features' ) 
final_weights   =   pd.merge(final_weights,   lasso_weights_df,   on= 'Features' ) 
 
 
```
